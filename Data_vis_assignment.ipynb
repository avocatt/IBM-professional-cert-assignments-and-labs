# Data Visualization with Python

## Peer Graded Assignment

A survey was conducted to gauge an audience interest in different data science topics, namely:

1. Big Data (Spark / Hadoop) 
2. Data Analysis / Statistics 
3. Data Journalism Data Visualization 
4. Data Visualization 
5. Deep Learning 
6. Machine Learning 

The participants had three options for each topic: Very Interested, Somewhat interested, and Not interested. 2,233 respondents completed the survey.


```python
import pandas as pd
```

```python
df = pd.read_csv("https://cocl.us/datascience_survey_data", index_col = 0)
df
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

```python
# 1. Sort the Very interestes in descending order.

df.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)

# 2. Convert the numbers into percentages of the total number of respondents.

df_pct = ((df / 2233) * 100).round(2)

ax = df_pct.plot(kind='bar',
                       figsize = (20, 8),
                       width = 0.8,
                       color = ['#5cb85c', '#5bc0de', '#d9534f'],
                       fontsize = 14)

plt.title('Percentage of Respondents Interests\''' in Data Science Areas', fontsize=16)


ax.set_facecolor((1.0, 1.0, 1.0))


y_axis = ax.axes.get_yaxis()
y_axis.set_visible(False)


for p in ax.patches:
    ax.annotate(str(p.get_height()) + '%', (p.get_x() * 1.005, p.get_height() * 1.03))
    
plt.show()
```
